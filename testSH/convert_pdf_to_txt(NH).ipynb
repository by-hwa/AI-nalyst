- 기업실적 예상치같이 표로 등장하는 자료는 변환 후 정리가 어렵습니다. "\n"이나 "\n\n"으로 분절 후 일정 길이를 넘는 것들만 문장으로 판단하고 합치려고 합니다. txt 변환 후 데이터셋에 포함될 문장 판단 기준에 대해 더 고민해보면 좋을 거 같습니다.
- 이메일, 이름 같은 개인정보는 전처리가 필요합니다. 그 외 추가적인 전처리를 적용할 필요가 있습니다.


```python
from collections import defaultdict


data_dict = defaultdict(list)
result_text = ""
label2id = {"buy": 0, "sell": 1, "hold": 2}
labels = []

for chunk in text.split("\n"):
    lower_chunk = chunk.lower()
    if "buy" in lower_chunk:
      labels.append("buy")
    if "sell" in lower_chunk:
      labels.append("sell")
    if "hold" in lower_chunk:
      labels.append("hold")
    
    if len(chunk) > 20:
      result_text += chunk

data_dict["text"].append(result_text)
data_dict["label"].append(label2id[labels[0]])
```

```python
from pdfminer.high_level import extract_text
import glob

pdfs = glob.glob("/content/drive/MyDrive/huggingface_course/*.pdf")
data_dict = defaultdict(list)
label2id = {"buy": 0, "sell": 1, "hold": 2}

for pdf in pdfs:
    text = extract_text(pdf)
    result_text = ""
    labels = []
    
    for chunk in text.split("\n"):
        lower_chunk = chunk.lower()
        if "buy" in lower_chunk or "매수" in lower_chunk:
          labels.append("buy")
        if "sell" in lower_chunk or "매도" in lower_chunk:
          labels.append("sell")
        if "hold" in lower_chunk or "유지" in lower_chunk or "보유" in lower_chunk:
          labels.append("hold")
        
        if len(chunk) > 20:
          result_text += chunk

    data_dict["text"].append(result_text)
    data_dict["label"].append(label2id[labels[0]])

result = pd.DataFrame(data_dict)
```
